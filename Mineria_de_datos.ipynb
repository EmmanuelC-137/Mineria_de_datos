#### Conceptos

##### La minería de datos o Data Mining, es el proceso técnico, automático o semi-automático, de extraer información oculta y patrones útiles a partir de grandes volúmenes de datos.

##### No se trata simplemente de "recopilar" datos, sino de analizarlos para transformar datos brutos en conocimiento accionable. Se sitúa en la intersección de la estadística, la inteligencia artificial y las bases de datos.
